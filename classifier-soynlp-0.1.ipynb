```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('전문가라벨링데이터.csv', encoding='utf-8')
```

```python
df.head()
```

```python
df_dr = df.drop(columns=['expert_comment_ko'], )
```

```python
df_dr
```

```python
from soynlp.utils import DoublespaceLineCorpus
from soynlp.noun import LRNounExtractor_v2
from soynlp.word import WordExtractor
from soynlp.tokenizer import LTokenizer
import re
```

```python
def preprocess_text(text):
    text = re.sub(r'[,\.\!\?\r\n]', '', text)
    return text.strip()
```

```python
sents = df['summary1'].apply(preprocess_text).to_list()
sents
```

```python
noun_extractor = LRNounExtractor_v2(verbose=True, extract_compound=True)
nouns = noun_extractor.train_extract(sents)
```

```python
"""
단어 점수 학습 함수
df - 학습할 문장이 있는 데이터프레임
text_column - 학습한 문장이 있는 열
"""
def wordscore_learn(df, text_column):

    text = df[text_column].apply(preprocess_text)

    sentences = text.tolist()

    word_extractor = WordExtractor(
        min_frequency=50,
        min_cohesion_forward=0.05,
        min_right_branching_entropy=0.0
    )
    word_extractor.train(sentences)

    word_scores = word_extractor.extract()

    return word_scores
```

```python
words = wordscore_learn(df_dr, 'summary1')
```

```python
cohesion_score = {word:score.cohesion_forward for word, score in words.items()}

noun_scores = {noun:score.score for noun, score in nouns.items()}
combined_scores = {noun:score + cohesion_score.get(noun, 0)
    for noun, score in noun_scores.items()}
combined_scores.update(
    {subword:cohesion for subword, cohesion in cohesion_score.items()
    if not (subword in combined_scores)}
)

tokenizer = LTokenizer(scores=combined_scores)
```
