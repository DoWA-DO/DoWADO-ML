```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer, util
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report

model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS') 
```

```python
df = pd.read_csv('merge_모든데이터.csv')
len(df)
```

```python
sentences = df['combined_conversation']
vectors = model.encode(sentences)
len(vectors[1]) 

vectors_df = pd.DataFrame(vectors)
vectors_df
```

```python
vectors_df[]
```

```python
job = pd.read_csv('labeling docs 직업목록.csv')
job = job['job_name'].copy()
```

```python
vectors_job = model.encode(job)
vectors_job
```

```python
similariries = util.cos_sim(vectors, vectors_job)
```

```python
similariries
```

```python
most_similar_items = np.argmax(similariries, axis=1)
most_similar_items
```

```python
#df.head(1)
len(most_similar_items)
```

```python
sim = most_similar_items.reshape(-1, 1)
```

```python
sim = pd.DataFrame(sim)
```

```python
sim
```

```python
df = pd.concat([df, sim], axis=1)
df
```

```python
job = pd.DataFrame(job)
job_dic = job['job_name'].to_dict()
```

```python
df['maped_job_name'] = df[0].map(job_dic)
```

```python
df
```

```python
ans = 0
for i, j in zip(df['priority1'], df['maped_job_name']):
    if i == j:
        ans += 1
ans
```

```python
vectors_df['job_name'] = df['priority1']
```

```python
job_dict2 = {job: index for index, job in enumerate(job['job_name'])}
job_dict2
```

```python
vectors_df['job_name'] = vectors_df['job_name'].map(job_dict2)
vectors_df
```

```python
x = vectors_df.drop(columns='job_name')
y = vectors_df[['job_name']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
print(classification_report(model.predict(x_test), y_test))
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
print(classification_report(model.predict(x_test), y_test))
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)
print(classification_report(model.predict(x_test), y_test))
```

```python
from lightgbm import LGBMClassifier

model = LGBMClassifier()
model.fit(x_train, y_train)
print(classification_report(model.predict(x_test), y_test))
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(x_train, y_train)
print(classification_report(model.predict(x_test), y_test))
```
